```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

v=np.linspace(0,1,500) #portion of vaccinated
v=np.array(v)

R_0=15
n_c=25
lambda_G=(R_0**2)/(R_0+n_c-1)

R=[]
for i in range(len(v)-1):
  R.append((1+(n_c-1)*(1-v[i]))*(1-v[i])*lambda_G)


y=np.ones(len(R))
for i in range(len(R)-1):
  if R[i]>=1 and R[i+1]<1:
    idx=i


Q=plt.figure(figsize=(10,8))
axes= Q.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlim([0.8,1])
axes.set_ylim([0,7])
plt.plot(v[0:len(v)-1],R,'k')
plt.plot(v[idx], R[idx], 'r*')
plt.xlabel('Fraction of vaccinated people v', fontsize=18)
plt.ylabel('$R_{*,e}$', fontsize=18)
#plt.title('Threshold clump parameter $R_{*,e}$')


#Now we fix v=v[idx] and we see what value of pi makes R_* be greater than one
print(v[idx])
pi=np.linspace(0,v[idx],500) #portion of schools with mandate, pi<v
pi=np.array(pi)

R1=[]
for i in range(len(pi)-1):
  R1.append((1+(n_c-1)*(1-v[idx])/(1-pi[i]))*(1-v[idx])*lambda_G)


P=plt.figure(figsize=(10,8))
axes= P.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlim([0,1])
axes.set_ylim([0,10])
plt.plot(pi[0:len(pi)-1],R1,'k')
plt.xlabel('Fraction of schools with mandate $\pi$',fontsize=18)
plt.ylabel('$R_{*,e}$',fontsize=18)
#plt.title('Threshold clump parameter $R_{*,e}$')
```
